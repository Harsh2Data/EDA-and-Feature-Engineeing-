# EDA and feature Engineering for Flight Price Prediction 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline   # writing that for print graphs without using plt.show 
```

```python
train_df = pd.read_excel('Data_train.xlsx')
train_df.head()
```

```python
test_df = pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
final_df=pd.concat([train_df,test_df],ignore_index=True)
final_df.head()
```

```python
final_df.tail()
```

```python
#feature engineering 
final_df['Date']= final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']= final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']= final_df['Date_of_Journey'].str.split('/').str[2]
final_df.head(2)
```

```python

final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
 final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)   
```

```python
final_df.head()
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
final_df.head()
```

```python
final_df['Arrival_Hour']=final_df['Arrival_Time'].apply(lambda x:x.split(':')[0])
final_df['Arrival_Minute']=final_df['Arrival_Time'].apply(lambda x:x.split(':')[1])
```

```python
final_df['Arrival_Hour']=final_df['Arrival_Hour'].astype(int)
final_df['Arrival_Minute']=final_df['Arrival_Minute'].astype(int)
```

```python
# final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df.head()
```

```python
final_df['Dep_Hour']=final_df['Dep_Time'].apply(lambda x:x.split(':')[0]).astype(int)
final_df['Dep_Minute']=final_df['Dep_Time'].apply(lambda x:x.split(':')[1]).astype(int)
```

```python
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
final_df.info()
```

```python
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Duration'].isnull().value_counts()
```

```python
final_df['duration_hrs']= final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
final_df['duration_hrs']


```

```python
final_df[final_df['duration_hrs']=='5m']
```

```python
final_df.drop(6474,axis = 0,inplace=True)
final_df.drop(13343,axis = 0,inplace=True)
```

```python
final_df['duration_hrs']=final_df['duration_hrs'].astype(int)
```

```python
final_df['duration_min']=final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]

final_df['duration_min'].fillna(0,inplace=True)
final_df['duration_min'].astype(int)

final_df.head()
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Duration_in_minutes'] = final_df['duration_hrs']*60
final_df['duration_min']= final_df['duration_min'].astype(int)
final_df['Duration_in_minutes']=final_df['Duration_in_minutes'].astype(int)+ final_df['duration_min'].astype(int)
final_df.head()
```

```python
final_df.drop('Duration',axis=1,inplace=True)
```
